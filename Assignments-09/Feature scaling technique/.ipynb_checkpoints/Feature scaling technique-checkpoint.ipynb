### Normalization / min max scaler

```python
max_alcohol = df.Alcohol.max()
min_alcohol = df.Alcohol.min()
```

```python
max_alcohol
```

```python
min_alcohol
```

```python
df['Alcohol_norm_manual'] = (df.Alcohol - min_alcohol ) / (max_alcohol - min_alcohol)
```

```python
df.filter(like='Alcohol')
```

```python
df['Alcohol']
```

# Now calculate normalization using sklearn 

```python
mmx = MinMaxScaler()
```

```python
df['alcohol_norm_sklearn'] = mmx.fit_transform(df[['Alcohol']])
```

```python
df.head()
```

# 2.Standardization

Manually calculation

```python
mean_alcohol = df.Alcohol.mean()
std = df.Alcohol.std()
```

```python
df['alcohol_std_manually'] = (df['Alcohol'] - mean_alcohol) / std
```

```python
df.head()
```

Standardization using sklearn

```python
scaler = StandardScaler()
```

```python
df['alcohol_std_sklearn'] = scaler.fit_transform(df[['Alcohol']])
```

```python
df.head()
```

### Robust Scaler (outliers)

```python
robust = RobustScaler()
```

```python
df['robust_alcohol'] = robust.fit_transform(df[['Alcohol']])
```

```python
df.head()
```

# Max Absolute Scaler

manually calculate

```python
max_abs = df.Alcohol.max()
```

```python
df['max_abs_manually'] = df.Alcohol / max_abs
```

```python
df.head()
```

Using sklearn

```python
mabs = MaxAbsScaler()
```

```python
df['max_abs_alcohol'] = mabs.fit_transform(df[['Alcohol']])
```

```python
df.head()
```

# Log Transformation

```python
fc = FunctionTransformer(np.log1p)
df['alcohol_fc'] = fc.fit_transform(df[['Alcohol']])
```

```python
df.head()
```

```python
df['fc_np'] = np.log1p(df[['Alcohol']]) # Another way
```

```python
df.head()
```

# Power Transformar

```python
log = PowerTransformer()
df['Alcohol_log'] = log.fit_transform(df[['Alcohol']])
```

```python
df['Alcohol_log'] = log.fit_transform(df[['Alcohol']])
```

```python
df.head()
```

### Power Transformer (box cox)

```python
from scipy.stats import boxcox
```

```python
df['box cox'], _ = boxcox(df['Alcohol'] + 1) #data,param
```

```python
df.head()
```

```python
df['Alcohol_log'] = np.log(df['Alcohol'] + 1) 
df['Alcohol_log'] = np.log10(df['Alcohol'] + 1)  # Logarithm base 10
```
